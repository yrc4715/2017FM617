```python
import requests
import pandas as pd
from datetime import datetime
import io

def get_stockprice(stock,startdate):
    url = "http://finance.google.com/finance/historical"
    params = {}
    params['q'] = stock
    params['startdate'] = startdate
    params['output'] = 'csv'
    
    r = requests.get(url, params=params)
    stockprice = pd.read_csv(io.StringIO(r.text)) 
    print(stockprice)
    
get_stockprice('AAPL',"2016-1-1")
```
